```python
import numpy
import pandas
import wvpy.util
```

```python

d = pandas.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [False, False, True, True, False]
})

wvpy.util.plot_roc(
    prediction=d['x'],
    istrue=d['y'],
    extra_points=pandas.DataFrame({
        'tpr': [0, 1],
        'fpr': [0, 1],
        'label': ['AAA', 'BBB']
    })
)
```

```python
d = pandas.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [False, False, True, True, False]
})

wvpy.util.dual_density_plot(
    probs=d['x'],
    istrue=d['y'],
)
```

```python
d = pandas.DataFrame({
    'x': [.1, .2, .3, .4, .5],
    'y': [False, False, True, True, False]
})

wvpy.util.dual_hist_plot(
    probs=d['x'],
    istrue=d['y'],
)
```

```python
d = pandas.DataFrame({
    'x': [.1, .2, .3, .4, .5],
    'y': [0, 0, 1, 1, 0]
})

wvpy.util.gain_curve_plot(
    prediction=d['x'],
    outcome=d['y'],
)
```

```python
d = pandas.DataFrame({
    'x': [.1, .2, .3, .4, .5],
    'y': [0, 0, 1, 1, 0]
})

wvpy.util.lift_curve_plot(
    prediction=d['x'],
    outcome=d['y'],
)
```

```python
d = pandas.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [False, False, True, True, False]
})

wvpy.util.threshold_plot(
    d,
    pred_var='x',
    truth_var='y',
    plotvars=("sensitivity", "specificity"),
)
```
