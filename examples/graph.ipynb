```python
import pandas
import wvpy.util

wvpy.__version__
```

```python
d = pandas.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [False, False, True, True, False]
})

d
```

```python
wvpy.util.threshold_plot(
    d,
    pred_var='x',
    truth_var='y',
    plotvars=("sensitivity", "specificity", "fraction"),
)
```

```python
wvpy.util.threshold_plot(
    d,
    pred_var='x',
    truth_var='y',
    plotvars=("precision", "recall", "fraction"),
)
```

```python
wvpy.util.threshold_plot(
    d,
    pred_var='x',
    truth_var='y',
    plotvars=("true_positive_rate", "false_positive_rate",
              "true_negative_rate", "false_negative_rate",),
)
```

```python
wvpy.util.threshold_plot(
    d,
    pred_var='x',
    truth_var='y',
    plotvars=("precision_misspelled", "recall"),
)
```
