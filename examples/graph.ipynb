```python
import pandas
import wvu.util

wvu.__version__
```

```python
d = pandas.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [False, False, True, True, False]
})

d
```

```python
wvu.util.threshold_plot(
    d,
    pred_var='x',
    truth_var='y',
    plotvars=("sensitivity", "specificity", "fraction"),
)
```

```python
wvu.util.threshold_plot(
    d,
    pred_var='x',
    truth_var='y',
    plotvars=("precision", "recall", "fraction"),
)
```

```python
wvu.util.threshold_plot(
    d,
    pred_var='x',
    truth_var='y',
    plotvars=("true_positive_rate", "false_positive_rate",
              "true_negative_rate", "false_negative_rate",),
)
```
